```python
def calculate_discounted_price(price):
    if price >= 5000: 
        return price * 0.75  
    elif price >= 2000: 
        return price * 0.50 
    elif price >= 1000: 
        return price * 0.25  
    else: 
        return price        

if __name__ == '__main__':
    yesorno = "Y" 
    
    while yesorno == "Y":
        try:
            price = float(input("Enter price: "))
            if price >= 0:
                final_price = calculate_discounted_price(price)
                print(f"Original Price: {price:.2f}")
                print(f"Final Price: {final_price:.2f}")
            else:
                print("Price cannot be negative.")
        except ValueError:
            print("Please enter a valid number.")
            
        yesorno = input("\nCalculate another? (Y or N): ")
```

```python
def factorial(a):
    fac = 1 
    while a > 0:
        fac *= a
        a -= 1
    return fac
def circle(r):
    return 3.14*r*r
def rectangle(l,b):
    return l*b
if __name__ == '__main__':
    yesorno = "Y" 
    while yesorno.upper() == "Y":
        try:
            
            a = int(input("Number: "))
            
            if a < 0:
                print("Please enter a positive integer or zero.")
            else:
                result = factorial(a)
                print(f"The factorial of {a} is: {result}")
                area=circle(a)
                print(f"The area of {a} is:{area}")
                
        except ValueError:
            print("Invalid input. Please enter a valid whole number.")
            
        yesorno = input("Do you want to calculate another? (Y/N): ")
```

```python
student_list = [
    {"name": "apple", "prn": "123"},
    {"name": "orange", "prn": "456"}
]

def presentornot(search_value, students_data):

    for student in students_data:
        if student.get("name") == search_value or student.get("prn") == search_value:
            return True
            
    return False


print(presentornot("apple", student_list)) 

print(presentornot("456", student_list))   

print(presentornot("grape", student_list))
```

positional and keyword are two type of parameters and their combination

```python
def example(x, y, z=30):
    return x, y, z
#positional arguementa
print(example(10, 20))

# Keyword arguments
print(example(x=10, y=20, z=380))

# Mixed positional + keyword arguments
print(example(10, y=720, z=30))
```
